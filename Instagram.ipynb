**Merge Engagement Data**

```python
import pandas as pd

# Load dataset
users = pd.read_csv("users.csv")
photos = pd.read_csv("photos.csv")
likes = pd.read_csv("likes.csv")
comments = pd.read_csv("comments.csv")

# Clean column names
users.columns = users.columns.str.strip().str.lower().str.replace(" ", "_")
photos.columns = photos.columns.str.strip().str.lower().str.replace(" ", "_")
likes.columns = likes.columns.str.strip().str.lower().str.replace(" ", "_")
comments.columns = comments.columns.str.strip().str.lower().str.replace(" ", "_")

# Create likes count
likes_count = likes.groupby("photo").size().reset_index(name="likes_count")

# Create comments count
comments_count = comments.groupby("photo_id").size().reset_index(name="comments_count")

# Merge data
photo_engagement = photos.merge(
    likes_count,
    left_on="id",
    right_on="photo",
    how="left"
)

photo_engagement = photo_engagement.merge(
    comments_count,
    left_on="id",
    right_on="photo_id",
    how="left"
)

# Fill missing values
photo_engagement['likes_count'] = photo_engagement['likes_count'].fillna(0)
photo_engagement['comments_count'] = photo_engagement['comments_count'].fillna(0)

photo_engagement['total_engagement'] = (
    photo_engagement['likes_count'] + photo_engagement['comments_count']
)

photo_engagement.head()
```
